# matplotlib实例4

```python
# coding=utf-8
import numpy as np
import matplotlib.pyplot as plt


def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)


t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.plot(t1, f(t1), ' bo', t2, f(t2), 'k:')
plt.plot(t2, np.cos(2 * np.pi * t2), 'r--')
plt.show()
```

# matplotlib实例5

```python
# coding=utf-8
import numpy as np
import matplotlib.pyplot as plt


def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)


t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure("2subplot")
plt.subplot(211)  # subplot（）中的参数分别指定了numrows、numcols、fignum
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2 * np.pi * t2), 'r--')
plt.show()
```

# matplotlib实例6

```python
# coding=utf-8
import numpy as np
import matplotlib.pyplot as plt
ax = plt.subplot(111)
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line = plt.plot(t, s, lw=2)
plt.annotate(
    'local max',
    xy=(2, 1),
    xytext=(3, 1.5),
    arrowprops=dict(facecolor='black', shrink=0.05),
)
plt.ylim(-2, 2)
plt.show()
```

# matplotlib实例7

```python
# coding=utf-8
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
# 导入宋体字体文件
font_set = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=15)
datax = [1, 2, 3, 4]
dataY = [1, 2, 3, 1]
# plot还有很多参数，可以查API修改，如颜色，虚线等
plt.plot(datax, dataY)
plt.title("绘制直线", fontProperties=font_set)
plt.xlabel("x轴", FontProperties=font_set)
plt.ylabel("轴", FontProperties=font_set)
plt.show()
```

# matplotlib实例8

```python
# coding=utf-8
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
# 导入宋体字体文件
font_set = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=15)
x = [0, 1, 2, 4, 5, 6]
y = [1, 2, 3, 2.1, 4, 1]
# 竖的条形图
plt.bar(x, y)
# pIt.barh（x，y）#横的条形图，注意，y坐标
plt.title("条形图", FontProperties=font_set)
plt.xlabel("x轴", FontProperties=font_set)
plt.ylabel("y轴", FontProperties=font_set)
plt.show()
```
